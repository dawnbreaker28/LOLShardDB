```python
total_count_result = site.cargo_client.query(
    tables="ScoreboardPlayers=SP",
    fields="COUNT(*)=total",
    where="SP.DateTime_UTC >= '2024-01-01 00:00:00' AND SP.DateTime_UTC <= '2024-12-31 23:59:59'"
)

total_count = int(total_count_result[0]['total'])
batch_size = 500
iterations = (total_count // batch_size) + (1 if total_count % batch_size != 0 else 0)
print(f'total iterations: {iterations}')

playerStats = []
for i in range(iterations):
    print(f'iteration: {i}')
    playerStats.extend(
        site.cargo_client.query(
            tables="ScoreboardPlayers=SP",
            fields="SP.Name, SP.GameId, SP.Kills, SP.Deaths, SP.Assists, SP.Champion, SP.Team",
            where="SP.DateTime_UTC >= '2024-01-01 00:00:00' AND SP.DateTime_UTC <= '2024-12-31 23:59:59'",
            offset=i * batch_size,
            limit=batch_size
        )
    )
    time.sleep(2)

with open('./player_stats_2424.pkl', 'wb') as file:
    pickle.dump(playerStats, file)
```

```python
with open('./roster_changes.pkl', 'rb') as pkl_file:
    roster_changes = pickle.load(pkl_file)
roster_changes
```

```python
active_memberships = {}

team_history = []

for change in roster_changes:
    player = change['Player']
    direction = change['Direction']
    team = change['Team']

    date_str = change['RosterChangeId'].split('_')[0]

    if direction == 'Join':
        active_memberships[(player, team)] = date_str

    elif direction == 'Leave':
        start_date = active_memberships.get((player, team))
        if start_date:
            team_history.append((player, team, start_date, date_str))
            del active_memberships[(player, team)]
        else:
            pass

for (player, team), start_date in active_memberships.items():
    team_history.append((player, team, start_date, None))

team_history
```

```python
with open('./team_history.pkl', 'wb') as file:
    pickle.dump(team_history, file)
```
